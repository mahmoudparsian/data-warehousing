# 🧑‍🏫 Instructor Solutions — GROUP BY & Window Functions

```
import duckdb
con = duckdb.connect()

con.execute("""
CREATE OR REPLACE TABLE sales AS
SELECT * FROM read_csv_auto('sales.csv');
""")
```

## ✅ Total revenue per country

```
SELECT country, SUM(sales_amount) AS revenue
FROM sales
GROUP BY country
ORDER BY revenue DESC;
```

## ✅ Total revenue per product

```
SELECT product_name, SUM(sales_amount) AS revenue
FROM sales
GROUP BY product_name
ORDER BY revenue DESC;
```

## ✅ Transactions per gender

```
SELECT gender, COUNT(*) AS cnt
FROM sales
GROUP BY gender;
```

## ✅ Revenue per age group

```
SELECT 
CASE 
 WHEN age < 30 THEN '18-29'
 WHEN age < 50 THEN '30-49'
 ELSE '50+'
END AS age_group,
SUM(sales_amount) AS revenue
FROM sales
GROUP BY age_group;
```

## ✅ Monthly revenue

```
SELECT 
EXTRACT(YEAR FROM transaction_date) AS year,
EXTRACT(MONTH FROM transaction_date) AS month,
SUM(sales_amount) AS revenue
FROM sales
GROUP BY year, month
ORDER BY year, month;
```

## ✅ Top 5 products

```
SELECT product_name, SUM(sales_amount) AS revenue
FROM sales
GROUP BY product_name
ORDER BY revenue DESC
LIMIT 5;
```

## ✅ Average order value per country

```
SELECT country, AVG(sales_amount) AS avg_order
FROM sales
GROUP BY country;
```

## ✅ Revenue per day of week

```
SELECT STRFTIME(transaction_date, '%A') AS day_of_week,
SUM(sales_amount) AS revenue
FROM sales
GROUP BY day_of_week;
```

## ✅ Revenue per quarter

```
SELECT 
EXTRACT(YEAR FROM transaction_date) AS year,
EXTRACT(QUARTER FROM transaction_date) AS quarter,
SUM(sales_amount) AS revenue
FROM sales
GROUP BY year, quarter;
```

## ✅ Country-product matrix

```
SELECT country, product_name, SUM(sales_amount) AS revenue
FROM sales
GROUP BY country, product_name;
```

## ✅ Top 3 products per country (window)

```
SELECT *
FROM (
 SELECT country, product_name, SUM(sales_amount) AS revenue,
        RANK() OVER (PARTITION BY country ORDER BY SUM(sales_amount) DESC) rnk
 FROM sales
 GROUP BY country, product_name
) t
WHERE rnk <= 3;
```
